## 💎 Diamond Geometry & Data Integrity Analysis

---

### 📏 The Formula Breakdown
The **Total Depth Percentage** is a crucial metric that relates the height of the diamond to its average width.

<div style="padding: 15px; border: 1px solid transparent; border-radius: 4px; color: #31708f; background-color: #d9edf7; border-color: #bce8f1;">
    <strong>Calculation:</strong>
    $$\text{Depth \%} = \frac{z}{\left(\frac{x + y}{2}\right)} \times 100 = \frac{200 \times z}{x + y}$$
</div>

**Variable Definitions:**
* <span style="color:red">**x**</span>: Length (mm)
* <span style="color:green">**y**</span>: Width (mm)
* <span style="color:blue">**z**</span>: Depth/Height (mm)

---

### ⚠️ The Mathematical Logic Failure
When $x, y, \text{ or } z$ are recorded as **0**, the data becomes logically "broken":

1. **Division by Zero:** If $x$ and $y$ are both $0$, the formula becomes $\frac{z}{0}$, which is **mathematically undefined**.
2. **Logical Inconsistency:** At Index `11935`, the `depth` is **63.3**, but $x, y, z$ are all **0.00**. 
    > **Conclusion:** The `depth` value was likely recorded correctly, but the raw dimensions ($x, y, z$) were lost or entered as placeholders.

---

### 🛠️ Data Recovery Logic
If a row has valid $x$ and $y$ values but $z$ is missing (0), we can **reconstruct** the physical depth using the reported percentage:



<div style="padding: 15px; border: 1px solid transparent; border-radius: 4px; color: #3c763d; background-color: #dff0d8; border-color: #d6e9c6;">
    <strong>Reconstruction Formula:</strong>
    $$z = \frac{\text{depth} \times (x + y)}{200}$$
</div>

<div style="border-left: 5px solid #1f4e79; background-color:lightgray; padding:12px 16px; border-radius:8px; font-family:Segoe UI, Roboto, Arial; color:#203a43; font-size:14px; line-height:1.6; margin-bottom:12px;">
So to calculate $z$ we should have the values $x$, $y$ and $depth$
</div>

```python
# Step 1 : remove recodrs with values for x or y = 0:
drops = df[(df['x'] == 0) | (df['y'] == 0)].index

# Drop the rows
df.drop(index=drops, inplace=True)

# Reset the index
df.reset_index(drop=True, inplace=True)
```

```python
# Step 2 : calculate the right value of z for the data :
df['z_calculated'] = (df['depth'] * (df['x'] + df['y'])) /200
df['z_calculated'] = np.round(df['z_calculated'],2)
```

```python
df.head(10)
```

```python
# query rows that not equals for z values:
df[(df['z'] != df['z_calculated']) & (df['z'] == 0)]
```

```python
df[df['z'] != df['z_calculated']].shape[0]
```

```python
# update the values of z :

# 1. Create a boolean mask (This returns True/False for each row)
mask = df['z'] != df['z_calculated']

# 2. Update the 'z' column for those specific rows
df.loc[mask, 'z'] = df.loc[mask, 'z_calculated']

# 3. Drop the helper column
df.drop(columns='z_calculated', inplace=True)
```

## Correlation Analysis 

Based on the correlation matrix provided, we can extract several statistically meaningful insights about the structural relationships in the dataset:

---

#### 1️⃣ **Price is primarily driven by Carat (Weight Effect)**

* **carat ↔ price = 0.92** → Very strong positive correlation.
* This confirms that **carat weight is the dominant predictor of price**.
* Economically, this aligns with diamond pricing theory: price increases non-linearly with mass due to rarity scaling.

**Implication for modeling:**
Carat will likely carry the highest feature importance in any regression model.

---

#### 2️⃣ **High Multicollinearity Among Size Dimensions (x, y, z)**

* x ↔ y = 0.97
* x ↔ z = 0.99
* y ↔ z = 0.99
* carat ↔ x = 0.98

These extremely high correlations indicate **strong linear dependency between physical dimensions and weight**.

**Interpretation:**

* Carat is mathematically derived from volume (which depends on x, y, z).
* Including all of them together in linear regression may cause multicollinearity issues (inflated variance, unstable coefficients).

**Modeling Recommendation:**

* For linear models → consider removing x, y, z or applying regularization (Ridge).
* For tree-based models → multicollinearity is less problematic.

---

#### 3️⃣ **Depth and Table Have Weak Relationship with Price**

* depth ↔ price = −0.01 (negligible)
* table ↔ price = 0.13 (very weak)

These geometric proportions appear to have minimal direct linear impact on price compared to weight.

**Interpretation:**
Price variation is far more sensitive to size than to proportion metrics.

---

#### 4️⃣ **Dimensional Variables Strongly Correlate with Price**

* x ↔ price = 0.89
* y ↔ price = 0.87
* z ↔ price = 0.88

This again reinforces that **physical size explains price variance**.

---

### 🔎 Overall Structural Insight

The dataset shows a **size-dominant pricing structure**:

* Weight (carat) and physical dimensions explain most of price variability.
* Depth and table contribute marginally.
* Strong multicollinearity exists among physical measurements.

---

### 📌 Final Modeling Guidance

If your objective is price prediction:

* ✔ Keep **carat** as a core predictor.
* ✔ Carefully treat **x, y, z** to avoid redundancy.
* ✔ Consider regularization if using linear regression.
* ✔ Tree-based models (XGBoost, LightGBM, CatBoost) can naturally handle correlated predictors.

```python
# Applying PCA on x,y and z :

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

dims = df[['x','y','z']]

ss = StandardScaler()
scaled_dims = ss.fit_transform(dims)

pca = PCA(n_components=1)
df['size_index'] = pca.fit_transform(scaled_dims)

print(f"Explained Variance: {pca.explained_variance_ratio_[0]:.2%}")
```


<div style="padding: 15px; border: 1px solid transparent; border-radius: 10px; background-color: lightgray; border-color: #d6e9c6;">
    <strong>Conclusion :</strong>
    The first Principal Component explains 99.6% of the variance in (x, y, z), indicating that diamond dimensions are highly redundant. We can effectively represent the diamond's physical size using a single feature (PC1) without significant loss of information.

```python
import scipy.stats as stats
```


<div style="border-left: 5px solid #1f4e79; background-color:lightgray; padding:12px 16px; border-radius:8px; font-family:Segoe UI, Roboto, Arial; color:#203a43; font-size:14px; line-height:1.6; margin-bottom:12px;">
    
<b>P-value</b> is too small <b>~ near to zero</b>

So There is a statistically <b>strong</b> correlation between  <b>color</b> and  <b>clarity</b>
</div>

```python
# 1. Prepare Tables in a list
contingency_tables = [
    pd.crosstab(df['color'], df['clarity']),
    pd.crosstab(df['color'], df['cut']),
    pd.crosstab(df['cut'], df['clarity'])
]

# 2. Variables' names for each table
table_names = [
    "color × clarity",
    "color × cut",
    "cut × clarity"
]

# 3. Create Cramér's V function
def cramers_v(confusion_matrix):
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

# 4. Testing on tables:
for table, name in zip(contingency_tables, table_names):
    chi2, p, dof, ex = stats.chi2_contingency(table)
    v = cramers_v(table)
    print(f"{name} | Chi2 p-value: {p:.4f} | Cramer's V: {v:.4f}")
```

<h3 style="color:#2E86C1;"> Conclusion from Chi-square & Cramér's V Analysis</h3>

- <span style="color: black;">**Interpretation:**</span>  
  - Although all relationships are statistically significant, the **practical strength of association is weak**.  
  - The strongest relationship is between **Cut and Clarity**, but still weak in practical terms.  
  - **Color** has very little practical effect on either **Clarity** or **Cut**.

- <span style="color: black;">**Implication for analysis:**</span>  
  - Focus on **Cut and Clarity** for predictive insights.  
  - **Color** can be considered largely independent in terms of predicting **Cut** or **Clarity**.

```python
# ANOVA: price across clarity

groups = [df.loc[df['clarity']==c, 'price'] for c in df['clarity'].unique()]
fstat, pval = stats.f_oneway(*groups)
print("ANOVA F:", fstat, "p:", pval)
```

```python
# ANOVA: price across cut

groups = [df.loc[df['cut']==c, 'price'] for c in df['cut'].unique()]
fstat, pval = stats.f_oneway(*groups)
print("ANOVA F:", fstat, "p:", pval)
```
